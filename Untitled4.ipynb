```python
import pandas as pd
from matplotlib import pyplot as plt
Data = {'country':['USA','canada','Germany','UK','France'],
        'GDP_Per_Capita':[45000,42000,53000,49000,47000],
        'Income_Per_capita':[4000,5000,6000,7000,8000]}
df = pd.DataFrame(Data)
df.plot(x = 'country',y=['GDP_Per_Capita','Income_Per_capita'],kind = 'bar')
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=[15,10])
totalDeath = [113055,37366,5973,35262]
totalRecovery = [778575,34256,230757,129404,365283]
activeCases = [1133536,4637363,686970,474855,89685]
country = ['USA','Brazil','rassia','India','Italy']
X= np.arange(len(totalDeath))

plt.bar(X,totalDeath,color = 'black',width = 0.25)
plt.bar(X + 0.25, totalRecovery, color = 'g',width = 0.25)
plt.bar(X + 0.5, activeCases, color = 'b',width = 0.25)

plt.legend(['Total Deaths','TotalRecovery','Active Cases'])
plt.xticks ([i + 0.25 for i in range(5)],country)

plt.title("Bar plot representing the total deapths,total recovered casaes and active cases country wise")

plt.xlabel('countries')
plt.ylabel('Cases')

plt.savefig('Barplot.png')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
labels = ['civil','Electrical','Mechanical','Chemical']
sizes = [15,50,45,10]
fig,ax = plt.subplots()
ax.pie(sizes,labels= labels,autopct = '%2.2f%%')
ax.axis('equal')
ax.set_title('Engeering Diciplines')
plt.show()
```

```python
from pandas import DataFrame
import matplotlib.pyplot as plt

Data = {'Tasks':[300,500,700],
        'Task Type': ['Tasks Pending','Tasks Ongoing','Tasks Completed']}
df = DataFrame(Data)
df.set_index('Task Type',inplace = True)
df.plot.pie(y = 'Tasks',figsize = (10,10),autopct = '%1.1f%%',startangle = 90)
```

```python
import numpy as np
import matplotlib.pyplot as plt 

labels = ['Civil', 'Electrical', 'Mechanical','Chemical']
sizes = [15, 20, 35, 89]

explode = (0.1, 0.1, 0.1, 0.4)

fig, ax = plt.subplots()
ax.pie(sizes,
       explode=explode,
       labels=labels, 
       autopct='%1.1f%%',
       shadow=True,
       startangle=90)
ax.axis('equal')
ax.set_title('Engineering Diciplines')

plt.show()
```
