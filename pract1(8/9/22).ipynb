```python
import pandas as pd
import numpy as np
```

```python
state=pd.read_csv("US_violent_crime.csv.csv")
state.head()
```

```python
state.transform(func = lambda x : x *10)
```

```python
mean_purchase = state.groupby('State')["Murder"].mean().rename("user_mean").reset_index()
print(mean_purchase)
```

```python
mer = state.merge(mean_purchase)
mer
```
